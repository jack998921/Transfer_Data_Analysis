```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import csv
import time
```

```python
transfer=[]
```

```python
def main_code():
    global transfer
    transfer = []
    
    chrome_options = Options()
    chrome_options.add_argument("--headless")
    chrome_options.add_argument("--disable-gpu")
    chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")
    
    driver = webdriver.Chrome(options=chrome_options)
    link = "https://www.yallakora.com/transfer-list/0/%D8%AC%D9%85%D9%8A%D8%B1-%D8%A7%D9%84%D8%A7%D9%86%D8%AA%D9%82%D8%A7%D9%84%D8%A7%D8%AA"
    driver.get(link)
    
    for i in range(1000):
        try:
            more_button = WebDriverWait(driver, 5).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, "a.moreBtn.icon-more"))
            )
            driver.execute_script("arguments[0].click();", more_button)
            time.sleep(1.5)
        except:
            break
            
    soup = BeautifulSoup(driver.page_source, "html.parser")
    driver.quit()
    
    data = soup.find_all("div", {"class": "wRow"})
    
    for i in data:
        date_tag = i.find('div', {'class': 'item fixed'})
        player_tag = i.find('div', {'class': 'item var'})
        from_tag = i.find('div', {'class': 'item var fromTeam'})
        to_tag = i.find('div', {'class': 'item var toTeam'})
        price_tag = i.find('div', {'class': 'item fixed price'})
        
        date = date_tag.text.strip() if date_tag else "unknown"
        player_name = player_tag.text.strip() if player_tag else "unknown"
        from_team = from_tag.text.strip() if from_tag else "unknown"
        to_team = to_tag.text.strip() if to_tag else "unknown"
        transfer_type = price_tag.text.strip() if price_tag else "unknown"
        
        transfer.append({
            'date': date,
            'player_name': player_name,
            'from_team': from_team,
            'to_team': to_team,
            'transfer_type': transfer_type
        })


        

```

```python
def printing():
    if not transfer:
        print("Empty")
        return
        
    header = transfer[0].keys()
    path = "C:/Users/KimoStore/Desktop/Transfer.csv"
    
    with open(path, 'w', encoding='utf-8-sig', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=header)
        writer.writeheader()
        writer.writerows(transfer)
        print("ok")
```

```python
main_code()
printing()
```
